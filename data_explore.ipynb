```python
def get_numeric_cols(df):
    # Define the prefixes to check for
    prefixes_to_check = ['int', 'float', 'uint']

    # Filter columns that start with the specified prefixes
    numeric_columns = [col for col in df.columns if any(col.startswith(prefix) for prefix in prefixes_to_check)]
    return numeric_columns
```

```python
colinfo_meta = []
num_cols = get_numeric_cols(meta_df)
for col in meta_df.columns:
    coltype = meta_df[col].dtype
    colinfo_meta.append({'column':col, 'type': str(coltype)})
```

```python
colinfo_meta
```

```python
colinfo_flow = []
num_cols = get_numeric_cols(flow_df)
for col in flow_df.columns:
    coltype = flow_df[col].dtype
    colinfo_flow.append({'column':col, 'type': str(coltype)})
```

```python
colinfo_flow
```

```python
pd.to_datetime(flow_df['Date']).astype(int)// 10**9 
```

```python
merged_df = pd.merge(meta_df, flow_df, on='Store', how='right')
```

```python
merged_df
```

## Change Datetime

```python
merged_df['Date'] =  pd.to_datetime(merged_df['Date'])
merged_df['Date'] = merged_df['Date'].astype(int)// 10**9
```

```python
merged_df['Date']
```

```python
sid = merged_df['Store'].unique()
print(sid.shape)
```

```python
sampled_df1 = pd.DataFrame(columns=merged_df.columns)
sampled_df2 = pd.DataFrame(columns=merged_df.columns)
for i, value in enumerate(sid):
    if i<100:
        sampled_df1 = pd.concat([sampled_df1, merged_df[merged_df['Store']==value]])
    elif i<200:
        sampled_df2 = pd.concat([sampled_df2, merged_df[merged_df['Store']==value]])
    else:
        break
```

```python
sampled_df1_= sampled_df1.drop('Id', axis=1)
sampled_df2_= sampled_df2.drop('Id', axis=1)
sampled_df1_.to_csv("./data/rss/sampled_merged_data1.csv", index=False)
sampled_df2_.to_csv("./data/rss/sampled_merged_data2.csv", index=False)
```
